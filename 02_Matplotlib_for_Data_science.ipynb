# Installation et importation du package

```python
#Importation des packages
import matplotlib.pyplot as plt
import numpy as np
```

# Tracé de graphiques simples avec Matplotlib

```python
# Génération de données
x=np.linspace(0,10,1000)
y=np.exp(x)
```

```python
# Tracer un graphique simple
plt.plot(x,y)
plt.show()
```


# Personnalisation de graphiques avec Matplotlib

```python
#Changer la taille d'un graphique avec la lageur et la hauteru de la figure en pouce
plt.figure()
plt.plot(x,y)
```

```python
# Titre et axes
plt.plot(x,y)
plt.title("Fonction exponentielle")
plt.xlabel("x")
plt.ylabel("Valeur")
plt.show()
```

```python
# Changement de Couleur
plt.plot(x,y,c="red")
plt.title("Fonction exponentielle")
plt.xlabel("x")
plt.ylabel("Valeur")
plt.show()
```

```python
# Changer le type de ligne
plt.plot(x,y,c="red",ls="--")
plt.title("Fonction exponentielle")
plt.xlabel("x")
plt.ylabel("Valeur")
plt.show()
```

```python
# Ajouter une Légende

plt.plot(x,y,c="red",ls="-",label="exp")
plt.title("Fonction exponentielle")
plt.xlabel("x")
plt.ylabel("Valeur")
plt.legend()
plt.show()
```

```python
# Ajouter une grille sur le graphique
plt.plot(x,y,c="red",ls="-",label="exp")
plt.title("Fonction exponentielle")
plt.xlabel("x")
plt.ylabel("Valeur")
plt.legend()
plt.grid()
plt.show()
```

# Faire plusieurs graphiques

## Faire des graphique sur un même carré

```python
# Créer un graphique vide
z=2*y
y3=z*2
plt.plot(x,y,c="red",ls="-",label="exp")
plt.plot(x,z,c="blue",ls="-",label="2*exp")
plt.plot(x,y3,c="green",ls="-",label="4*exp")
plt.title("Fonction exponentielle")
plt.xlabel("x")
plt.ylabel("Valeur")
plt.legend()
plt.show()
```

```python
# Changer les limites des axes
z=2*y
y3=z*2
plt.plot(x,y,c="red",ls="-",label="exp")
plt.plot(x,z,c="blue",ls="-",label="2*exp")
plt.plot(x,y3,c="green",ls="-",label="4*exp")
plt.xlim((6,10))
plt.ylim((1000,80000))
plt.title("Fonction exponentielle")
plt.xlabel("x")
plt.ylabel("Valeur")
plt.legend()
plt.show()
```

## Faire des subplots

### Approche fonctionnelle

```python
# Graphique avec subplot
plt.subplot(2,2,1)
plt.plot(x,y,c="red",ls="-",label="exp")
plt.subplot(2,2,2)
plt.plot(x,y,c="blue",ls="-",label="exp")
plt.subplot(2,2,3)
plt.plot(x,y,c="green",ls="-",label="exp")
plt.subplot(2,2,4)
plt.plot(x,y,c="black",ls="-",label="exp")
plt.subplots_adjust(wspace=0.5)
```

###  Approche orientée objet

```python
# Subplots
fig,axes=plt.subplots(2,1)
axes[0].plot(x,y,c="red")
axes[1].plot(x,y,c="blue")
```

# Sauvegarder le graphique 


```python
# Sauvegarder son graphique dans un fichier
plt.subplot(2,2,1)
plt.plot(x,y,c="red",ls="-",label="exp")
plt.subplot(2,2,2)
plt.plot(x,y,c="blue",ls="-",label="exp")
plt.subplot(2,2,3)
plt.plot(x,y,c="green",ls="-",label="exp")
plt.subplot(2,2,4)
plt.plot(x,y,c="black",ls="-",label="exp")
plt.subplots_adjust(wspace=0.5)
plt.savefig("figure.png")
plt.savefig("figure.jpeg")
```
